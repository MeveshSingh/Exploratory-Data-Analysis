```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('Train.csv')
dataset.head()
```

```python
dataset.shape
```

```python
#Missing Values
features_with_na=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1]

for feature in features_with_na:
    print(feature,np.round(dataset[feature].isnull().mean(),4), '%missing values')
```

```python
for feature in features_with_na:
    data=dataset.copy()
    data[feature]=np.where(data[feature].isnull(),1,0)
    
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

```python
print(f'Id of Houses:{len(dataset.Id)}')
```

```python
#Numerical Variables
numerical_feature=[feature for feature in dataset.columns if dataset[feature].dtypes!='O']
print(f'Number of numerical variables: {len(numerical_feature)}')
```

```python
dataset[numerical_feature].head()
```

```python
#Temporal Variable
year_feature=[feature for feature in numerical_feature if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
for feature in year_feature:
    print(feature,dataset[feature].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel(feature)
plt.ylabel('Sale Price')
plt.title('House price vs YearSold')
```

```python
for feature in year_feature:
    if feature!='YrSold':
        data=dataset.copy()
        
        data[feature]=data['YrSold']-data[feature]
        
        plt.scatter(data[feature],data["SalePrice"])
        plt.xlabel(feature)
        plt.ylabel("Sale Price")
        plt.show()
```

```python
discrete_feature=[feature for feature in numerical_feature if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]
print(f'Discrete variables count: {len(discrete_feature)}')
```

```python
discrete_feature
```

```python
dataset[discrete_feature].head()
```

```python
for feature in discrete_feature:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Sale Price')
    plt.title(feature)
    plt.show()
```

```python
continuous_feature=[feature for feature in numerical_feature if feature not in discrete_feature+year_feature+['Id']]
print(f'Continuous feature count: {len(continuous_feature)}')
```

```python
for feature in continuous_feature:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```
